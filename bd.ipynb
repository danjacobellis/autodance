```python
import bjontegaard as bd
import matplotlib.pyplot as plt
```

```python
r_webp = [1/840, 1/400, 1/200, 1/100, 1/50]
d_webp = [25.5, 28.5, 30.5, 33.5, 36.5]
r_cosmos = [1/384, 1/96]
d_cosmos = [26.5, 30.5]
r_ours = [1/160, 1/60]
d_ours= [31.5, 35.7]
```

```python
plt.plot(r_webp,d_webp,marker='.')
plt.plot(r_cosmos,d_cosmos,marker='.')
plt.plot(r_ours,d_ours,marker='.')
# plt.plot(r2,d2)
```

```python
bd_rate = bd.bd_rate(r_webp, d_webp, r_cosmos, d_cosmos, method='pchip', require_matching_points=False)
print(f"BD-Rate: {bd_rate:.4f} %")
```

```python
bd_rate = bd.bd_rate(r_webp, d_webp, r_ours, d_ours, method='pchip', require_matching_points=False)
print(f"BD-Rate: {bd_rate:.4f} %")
```

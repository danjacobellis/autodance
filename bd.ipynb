```python
import bjontegaard as bd
import matplotlib.pyplot as plt
```

```python
r_webp = [1/258.1, 1/142.79, 1/119.24, 1/93.34, 1/56.73, 1/41.43, 1/26.96]
d_webp = [28.33, 30.94, 31.77, 32.93, 35.44, 37.08, 39.48]
r_cosmos = [1/384, 1/96]
d_cosmos = [26.5, 30.5]
r_ours = [1/166.45, 1/38.02]
d_ours= [31.37, 38.88]
```

```python
plt.figure(figsize=(4,3),dpi=200)
plt.plot(r_webp,d_webp,marker='.',label='webp')
plt.plot(r_cosmos,d_cosmos,marker='.',label='cosmos (128x,32x)')
plt.plot(r_ours,d_ours,marker='.',label='ours (64x,16x)')
plt.legend()
plt.ylabel("PSNR")
plt.xlabel("Rate")
plt.grid()
```

```python
bd_rate = bd.bd_rate(r_webp, d_webp, r_cosmos, d_cosmos, method='pchip', require_matching_points=False)
print(f"BD-Rate: {bd_rate:.4f} %")
```

```python
bd_rate = bd.bd_rate(r_webp, d_webp, r_ours, d_ours, method='pchip', require_matching_points=False)
print(f"BD-Rate: {bd_rate:.4f} %")
```
